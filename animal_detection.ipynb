```python
import os
import random
import shutil

# Paths
img_dir = "/content/dataset/train/images"
label_dir = "/content/dataset/train/labels"
valid_img_dir = "/content/dataset/valid/images"
valid_label_dir = "/content/dataset/valid/labels"

# Create validation folders if not exist
os.makedirs(valid_img_dir, exist_ok=True)
os.makedirs(valid_label_dir, exist_ok=True)

# Get all image files
image_files = [f for f in os.listdir(img_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Shuffle and select 10% for validation
random.shuffle(image_files)
num_valid = int(len(image_files) * 0.1)  # 10% for validation
valid_images = image_files[:num_valid]

# Move images and corresponding labels
for img in valid_images:
    img_path = os.path.join(img_dir, img)
    label_path = os.path.join(label_dir, img.replace(img.split('.')[-1], 'txt'))  # Replace extension with .txt

    if os.path.exists(label_path):  # Move only if label exists
        shutil.move(img_path, valid_img_dir)
        shutil.move(label_path, valid_label_dir)

print(f"Moved {num_valid} images and labels to validation set.")
```
